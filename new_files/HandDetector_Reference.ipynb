```python
import scipy.io as sio
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
from matplotlib import rcParams
rcParams["figure.figsize"] = 20, 15
```

```python
path_label = os.getcwd() + '\\images\\test\\CARDS_COURTYARD_B_T\\polygons.mat'
path_img = os.getcwd() + '\\images\\test\\CARDS_COURTYARD_B_T\\frame_0011.jpg'
mat = sio.loadmat(path_label)
img = np.array(mpimg.imread(path_img))
```

```python
mat.keys()
```

```python
pols = mat["polygons"][0]
pols.shape
```

```python
for point in pols[0]:
    if len(point) > 1:
        x = [n[0] for n in point]
        y = [n[1] for n in point]
        max_y = int(max(y))
        min_y = int(min(y))
        max_x = int(max(x))
        min_x = int(min(x))
        
        img = cv2.rectangle(img, (min_x, max_y), (max_x, min_y), (255, 0, 0), 2)
        
        for p in point:
            img = cv2.circle(img, (int(p[0]), int(p[1])), 1, (0, 255, 0), -1)
plt.imshow(img)
plt.show()
```
